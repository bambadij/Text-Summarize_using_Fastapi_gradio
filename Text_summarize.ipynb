```python
def summarize_text_pegasus(text, max_length=200):
    tokenizer = PegasusTokenizer.from_pretrained("google/pegasus-xsum")
    model = PegasusForConditionalGeneration.from_pretrained("google/pegasus-xsum")

    inputs = tokenizer([text], truncation=True, padding="longest", max_length=512, return_tensors="pt")

    with torch.no_grad():
        outputs = model.generate(inputs["input_ids"], num_beams=2, early_stopping=True,do_sample=False)

    summary_text = tokenizer.decode(outputs[0], skip_special_tokens=False)

    return summary_text

interface = gr.Interface(fn=summarize_text_pegasus, inputs="textbox",outputs="textbox", title="Text Summarizer Using Pegasus")
interface.launch(share=True)
```

```python
def summarize_text_output(text, max_length=200):
    tokenizer = AutoTokenizer.from_pretrained("almanach/camembert-base")
    config = AutoConfig.from_pretrained("almanach/camembert-base")
    config.is_decoder = True
    model = CamembertForCausalLM.from_pretrained("almanach/camembert-base", config=config)
    inputs = tokenizer([text], truncation=True, padding="longest", max_length=512, return_tensors="pt")

    with torch.no_grad():
        outputs = model.generate(inputs["input_ids"], max_length=max_length, num_beams=2, early_stopping=True)

    summary_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return summary_text

interface = gr.Interface(fn=summarize_text_output, inputs="textbox", outputs="textbox", title="Text Summarizer Using Camembert")
interface.launch(share=True)
```

```python
import transformers
from transformers import pipeline

def summarize_text_bart_(text, max_length=200):
    """Summarizes the input text and returns the summary along with word counts.

    Args:
        text (str): The input text to be summarized.
        max_length (int, optional): The maximum length of the generated summary.
            Defaults to 200.

    Returns:
        tuple: A tuple containing three elements:
            - summary_text (str): The generated summary of the input text.
            - input_word_count (int): The number of words in the input text.
            - output_word_count (int): The number of words in the generated summary.
    """

    # Ensure BART model is downloaded
    if not transformers.AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-large-cnn", local_files_only=True):
        transformers.AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-large-cnn")

    summarizer = pipeline('summarization', model="facebook/bart-large-cnn")

    summary = summarizer(text, max_length=max_length, min_length=30, do_sample=False)
    summary_text = summary[0].get('summary_text')

    # Count words in input text
    input_word_count = len(text.split())

    # Count words in output summary
    output_word_count = len(summary_text.split())

    return summary_text, input_word_count, output_word_count

interface = gr.Interface(fn=summarize_text_bart_, inputs="textbox", outputs="text", title="Text Summarizer Using BART with Word Counts",
                         elem_fn=lambda inputs, outputs: f"**Input Word Count:** {outputs[1]}\n**Output Word Count:** {outputs[2]}\n\n**Summary:** {outputs[0]}",
                         label="Summary")
interface.launch()
```
